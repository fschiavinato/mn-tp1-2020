## Testeo

```python

results, reference = utils.results(with_matches=False)
by_T = (lambda results: {(T, P): [result for file,result in results.items() if re.match(f'.*/random{T}_{P}/.*', file)] for T in Ts for P in Ps} )
results = by_T(results)
reference = by_T(reference)
error = {(T, P): [max(abs(((np.array(own)-np.array(ref))/np.array(own)))) for own, ref in zip(results[(T, P)], reference[(T, P)])] for (T, P) in results}
```
